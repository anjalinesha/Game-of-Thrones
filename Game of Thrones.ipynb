```python
import numpy as np
import pandas as pd
```

```python
battle = pd.read_csv("battles.csv")
```

```python
battle.head()
```

```python
battle.shape
```

```python
battle.rename(columns={'attacker_1':'primary_attacker'},inplace=True)
battle.head()
```

```python
battle.rename(columns={'defender_1':'primary_defender'},inplace=True)
battle.head()
```

```python
battle['attacker_king'].value_counts()
```

```python
battle['defender_king'].value_counts()
```

```python
battle['location'].value_counts()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.set(rc={'figure.figsize':(13,7)})
sns.barplot(x='attacker_king',y='attacker_size',data=battle)
plt.show()
```

Stannis Baratheon  has a huge army size 

```python
sns.set(rc={'figure.figsize':(13,7)})
sns.barplot(x='defender_king',y='defender_size',data=battle)
plt.show()
```

Renly Baratheon has a huge army size(in defence)

```python
#to know wht are battle types that attackers did
sns.countplot(x=battle['attacker_king'],hue=battle['battle_type']) #hue=colour of the bars
plt.show
```
